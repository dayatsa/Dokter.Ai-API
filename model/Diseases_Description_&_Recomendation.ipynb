```python
import pandas as pd
import numpy as np
from googletrans import Translator
```

```python
translator = Translator()
x = translator.translate('안녕하세요.')
print(x)
```

```python
df1 = pd.read_csv("/content/B21-CAP0047-Dokter.ai-MachineLearningRepository/dataset/desease_based_symptoms_with_id.csv")
df1 = df1.set_index("Desease")
df1
```

```python
df2 = pd.read_csv("/content/B21-CAP0047-Dokter.ai-MachineLearningRepository/kaggle_dataset/dataset_2/symptom_precaution.csv")
df2 = df2.set_index("Disease")
df2
```

```python
df3 = pd.read_csv("/content/B21-CAP0047-Dokter.ai-MachineLearningRepository/kaggle_dataset/dataset_2/symptom_Description.csv")
df3 = df3.set_index("Disease")
df3
```

## Combine Data Frame

```python
df_combine = df1.merge(df2, how='left', left_index=True, right_index=True)
df_combine = df_combine.merge(df3, how = 'left', left_index=True, right_index=True)
df_combine.loc['Diabetes ', 'Description'] = df3.loc['Diabetes']['Description']
df_combine.loc['Hypertension ', 'Description'] = df3.loc['Hypertension']['Description']
df_combine.loc['Dimorphic hemmorhoids(piles)', 'Description'] = df3.loc['Dimorphic hemorrhoids(piles)']['Description']

df_combine
```

# Find All Recomendation

```python
rec_1 = list(df_combine['Precaution_1'].unique())
rec_2 = list(df_combine['Precaution_2'].unique())
rec_3 = list(df_combine['Precaution_3'].unique())
rec_4 = list(df_combine['Precaution_4'].unique())
rec_total = rec_1+rec_2+rec_3+rec_4
len(rec_total)
```

#Save Recomendation Before Translate to TXT File

```python
with open("Recomendation.txt", "w") as file:
  for rec in rec_total:
    try:
      file.write(rec)
      file.write('\n')
    except:
      pass
```

#Load Recomendation After Translate from TXT File to Dict

```python
with open("Recomendation.txt", "r") as file:
  list_rec = file.readlines()
dict_rec = {}
for rec in list_rec:
  rec = rec.rstrip("\n")
  rec = rec.split(', ')
  key = rec[0]
  value = rec[1]
  dict_rec[key] = value
```

# Combine

```python
for i in list(df_combine.index):
  recomendation = []
  for j in range(1,5):
    column_key = 'Precaution_' + str(j)
    if(df_combine.loc[i][column_key]!=np.NaN):
      key_dict = df_combine.loc[i][column_key]
      try:
        recomendation.append(dict_rec[key_dict])
      except:
        pass
    df_combine.at[i, "Recomendation"] = ", ".join(recomendation)
```

```python
df_combine
```

```python
df_combine = df_combine.reset_index()
df_combine
```

# Translate Description Diseases

```python
for i in list(df_combine.index):
  text_en = df_combine.loc[i]['Description']
  text_id = translator.translate(text_en, dest='id', src='en').text
  df_combine.at[i, "Description_id"] = text_id
```

```python
for i in list(df_combine.index):
  text_en = df_combine.loc[i]['Desease']
  text_id = translator.translate(text_en, dest='id', src='en').text
  df_combine.at[i, "Desease_id"] = text_id
```

```python
df_combine.set_index('desease_id')
df_combine.drop(columns=['Unnamed: 0','Precaution_1','Precaution_2','Precaution_3','Precaution_4','Symptoms','symptoms_id','Description']).reset_index().set_index('desease_id')
data_final = df_combine[['desease_id','Desease_id', 'Description_id', 'Recomendation']]
data_final = data_final.rename(columns={'desease_id' : 'ID', 'Desease_id':'Disease', 'Description_id':'Description'})
data_final = data_final.set_index('ID')
data_final['Image'] = 'Google.com'
data_final
```
