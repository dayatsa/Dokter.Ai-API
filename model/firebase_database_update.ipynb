```python
import pandas as pd
import requests
```

#Symptoms Update Firebase

```python
df_symptoms = pd.read_csv('/content/B21-CAP0047-Dokter.ai-MachineLearningRepository/dataset/Symptoms_Document.csv')
df_symptoms.head()
```

```python
def postSymptomps(id):
  json_data = {
    "image": df_symptoms['link gambar'][id],
    "name": df_symptoms['Gejala (Indonesia)'][id],
    "description": df_symptoms['Penjelasan'][id],
    "question" : df_symptoms['Bentuk Pertanyaan'][id]
  }

  response = requests.post('https://us-central1-dokterai-917bb.cloudfunctions.net/DokterAIAPI/symptom/create/'+str(id), json=json_data)
  # print(json_data)
  print("Status code: ", response.status_code)
  print(response.text)

  if response.text == 'Document is exist, use PATCH to update document!':
    updateSymptomps(id)


def updateSymptomps(id):
  json_data = {
    "image": df_symptoms['link gambar'][id],
    "name": df_symptoms['Gejala (Indonesia)'][id],
    "description": df_symptoms['Penjelasan'][id],
    "question" : df_symptoms['Bentuk Pertanyaan'][id]
  }

  response = requests.patch('https://us-central1-dokterai-917bb.cloudfunctions.net/DokterAIAPI/symptom/update/'+str(id), json=json_data)
  # print(json_data)
  print("Status code: ", response.status_code)
  print(response.text)
```

```python
for i in range(len(df_symptoms)):
  print("Post symptoms id : {}".format(i))
  postSymptomps(i)
  print()
```

#Disease Update Firebase

```python
df_diseases = pd.read_csv('/content/B21-CAP0047-Dokter.ai-MachineLearningRepository/dataset/Diseases_Document (1).csv')
df_diseases.head()
```

```python
def postDiseases(id):
  json_data = {
    "recomendation": df_diseases['Recomendation'][id].split(', '),
    "image": df_diseases['Image'][id],
    "name": df_diseases['Disease'][id],
    "description": df_diseases['Description'][id]
  }

  response = requests.post('https://us-central1-dokterai-917bb.cloudfunctions.net/DokterAIAPI/disease/create/'+str(id), json=json_data)
  # print(json_data)
  print("Status code: ", response.status_code)
  print(response.text)

  if response.text == 'Tidak dapat menambahkan data karena sudah ada, gunakan patch':
    updateDiseases(id)


def updateDiseases(id):
  json_data = {
    "recomendation": df_diseases['Recomendation'][id].split(', '),
    "image": df_diseases['Image'][id],
    "name": df_diseases['Disease'][id],
    "description": df_diseases['Description'][id]
  }

  response = requests.patch('https://us-central1-dokterai-917bb.cloudfunctions.net/DokterAIAPI/disease/update/'+str(id), json=json_data)
  # print(json_data)
  print("Status code: ", response.status_code)
  print(response.text)
```

```python
for i in range(len(df_diseases)):
  print("Post Diseases id : {}".format(i))
  postDiseases(i)
  print()
```
